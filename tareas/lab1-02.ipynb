```python
import numpy as np
```

```python
A = np.array([[1.32, 3.25, 2.77],
                   [-3.57, 2.12, -3.26],
                   [-9.87, 3.33, 2.15]])
```

#### Calcular la inversa de la matriz A empleando inv de linal y multiplicar A por su inversa ####

```python
invA = np.linalg.inv(A)
invA
```

```python
product = np.matmul(invA, A)
product
```

### Hacer lo mismo con 5 matrices distintas ###

```python
def multiply_matrices_A_and_invA(A):
    invA = np.linalg.inv(A)
    product = np.matmul(invA, A)
    return product

matrices =[
        np.array([[1.32, 3.25, 2.77],
                   [-3.57, 2.12, -3.26],
                   [-9.87, 3.33, 2.15]]),
        np.array([[1.32, 3.25, 2.77],
                   [-3.57, 2.12, -3.26],
                   [-9.87, 3.33, 2.15]]),
        np.array([[1.32, 3.25, 2.77],
                   [-3.57, 2.12, -3.26],
                   [-9.87, 3.33, 2.15]]),
        np.array([[1.32, 3.25, 2.77],
                   [-3.57, 2.12, -3.26],
                   [-9.87, 3.33, 2.15]]),
        np.array([[1.32, 3.25, 2.77],
                   [-3.57, 2.12, -3.26],
                   [-9.87, 3.33, 2.15]]),
    ]

for i, matrix in enumerate(matrices):
    print(f"Matrix {i+1}")
    print(multiply_matrices_A_and_invA(matrix))
    print("--------------------------------")
```
