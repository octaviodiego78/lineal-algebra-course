Jacobi y Gauss-Seidel

```python
import numpy as np
```

```python
def jacobi(A, b, tol, max_iter=100):
    n = len(b)
    x = np.zeros(n)
    
    for k in range(max_iter):
        x_new = np.zeros(n)
        
        for i in range(n):
            suma = sum(A[i][j] * x[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - suma) / A[i][i]
        
        norma_actual = sum(abs(x_new))
        norma_anterior = sum(abs(x))
        
        if norma_anterior != 0:
            cambio = abs((norma_actual - norma_anterior) / norma_anterior) * 100
            if cambio < tol:
                return x_new, k + 1
        
        x = x_new.copy()
    
    return "No converge", max_iter
```

```python
def gauss_seidel(A, b, tol, max_iter=100):
    n = len(b)
    x = np.zeros(n)
    
    for k in range(max_iter):
        x_anterior = x.copy()
        
        for i in range(n):
            suma = sum(A[i][j] * x[j] for j in range(n) if j != i)
            x[i] = (b[i] - suma) / A[i][i]
        
        norma_actual = sum(abs(x))
        norma_anterior = sum(abs(x_anterior))
        
        if norma_anterior != 0:
            cambio = abs((norma_actual - norma_anterior) / norma_anterior) * 100
            if cambio < tol:
                return x, k + 1
        
    return "No converge", max_iter
```

```python
A = np.array([[4.0, 1.0, -1.0], 
                [1.0, -3.0, 1.0], 
                [1.0, 2.0, 5.0]])
b = np.array([6.0, 2.0, 11.0])
tolerancia = 0.00001
```

```python
print("Método de Jacobi:")
solucion, iteraciones = jacobi(A, b, tolerancia)
print(f"Solución: {solucion}")
print(f"Iteraciones: {iteraciones}")
```

```python
print("\nMétodo de Gauss-Seidel:")
solucion, iteraciones = gauss_seidel(A, b, tolerancia)
print(f"Solución: {solucion}")
print(f"Iteraciones: {iteraciones}")
```
