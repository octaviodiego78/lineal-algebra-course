```python
import math
```

```python
def biseccion(f, a, b, tol=1e-6, max_iter=100):
    if f(a) * f(b) >= 0:
        return None
    
    for i in range(max_iter):
        c = (a + b) / 2
        if abs(f(c)) < tol or (b - a) / 2 < tol:
            return c
        if f(a) * f(c) < 0:
            b = c
        else:
            a = c
    return c
```

```python
def punto_fijo(g, x0, tol=1e-6, max_iter=100):
    for i in range(max_iter):
        x1 = g(x0)
        if abs(x1 - x0) < tol:
            return x1
        x0 = x1
    return x0
```

```python
def newton(f, df, x0, tol=1e-6, max_iter=100):
    for i in range(max_iter):
        fx = f(x0)
        if abs(fx) < tol:
            return x0
        dfx = df(x0)
        if abs(dfx) < 1e-12:
            return None
        x1 = x0 - fx / dfx
        if abs(x1 - x0) < tol:
            return x1
        x0 = x1
    return x0
```

```python
def f1(x):
    return x**3 - x - 1

def g1(x):
    return (x + 1)**(1/3)

def f2(x):
    return x**2 - 2

def df2(x):
    return 2*x

def f3(x):
    return math.exp(x) - 3*x

def df3(x):
    return math.exp(x) - 3
```

```python
print("Bisección - x³ - x - 1 = 0 en [1, 2]:")
result1 = biseccion(f1, 1, 2)
print(f"Raíz: {result1}")
```

```python
print("\nPunto fijo - x = (x+1)^(1/3), x₀ = 1.5:")
result2 = punto_fijo(g1, 1.5)
print(f"Raíz: {result2}")
```

```python
print("\nNewton - e^x - 3x = 0, x₀ = 1:")
result3 = newton(f3, df3, 1)
print(f"Raíz: {result3}")
```
