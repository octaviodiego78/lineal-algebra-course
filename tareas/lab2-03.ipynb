Factorizacion LU

```python
import numpy as np
```

```python
def lu_factorization(A):
    A = np.array(A, dtype=float)
    n = len(A)
    L = np.eye(n)
    U = A.copy()

    # Creando la triangular superior
    for i in range(n):
        if U[i][i] == 0:
            return "No se puede factorizar"
        
        # Eliminando la columna debajo de la diagonal
        for j in range(i + 1, n):
            multiplier = U[j][i] / U[i][i]
            L[j][i] = multiplier 
            U[j][i:] -= multiplier * U[i][i:]
    return L, U
```

```python
A = [
        [0, 2, 1],
        [1, -2, -3],
        [3, -1, 1]
    ]

r = lu_factorization(A)
print(r)
```

```python
A = [
        [2, 2, 1],
        [1, -2, -3],
        [3, -1, 1]
    ]

l,u = lu_factorization(A)
print("L:")
print(l)
print("--------------------------------")
print("U:")
print(u)
```
