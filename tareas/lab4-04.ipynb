```python
import numpy as np
from numpy import linalg as LA
```

```python
def descomposicion_svd(A):
    m, n = A.shape
    AtA = A.T @ A
    eigenvalues, V = LA.eig(AtA)
    
    singular_values = np.sqrt(np.abs(eigenvalues))
    idx = np.argsort(singular_values)[::-1]
    singular_values = singular_values[idx]
    V = V[:, idx]
    
    U = A @ V
    for i in range(min(m, n)):
        if singular_values[i] > 1e-10:
            U[:, i] = U[:, i] / singular_values[i]
    
    return U[:, :min(m, n)], singular_values[:min(m, n)], V.T
```

```python
A = np.random.rand(8, 5)
    
U, s, Vt = descomposicion_svd(A)
U_numpy, s_numpy, Vt_numpy = LA.svd(A, full_matrices=False)
```
