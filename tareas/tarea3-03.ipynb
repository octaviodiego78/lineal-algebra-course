### Investigue el método de Broyden para resolver sistemas de ecuaciones no lineales y compare la convergencia con el método de Newton. ###

El metodo de Broyden es quasi-newtoniano, ya que es muy parecido al metedo de Newton, sin embargo Broyden:

- Aproxima la Jacobiana en vez de calcularla en cada iteracion
- En cada iteracion actualiza la aproximacion de la jacobiana
- Tiene convergencia superlineal, mayor que Newtown(cuadratica), sin embargo el coste es menor ya que se evita calcular la Jacobiana
