```python
import numpy as np
from numpy import linalg as LA
```

```python
def metodo_qr(A, tol=1e-7, max_iter=1000):
    n = A.shape[0]
    Ak = A.copy()
    Q_total = np.eye(n)
    
    for _ in range(max_iter):
        Q, R = np.linalg.qr(Ak)
        Ak_new = R @ Q
        Q_total = Q_total @ Q
        
        if np.linalg.norm(Ak_new - Ak) < tol:
            break
        Ak = Ak_new
    
    eigenvalues = np.diag(Ak)
    eigenvectors = Q_total
    
    return eigenvalues, eigenvectors
```

```python
A = np.array([[8, 3, 1, 7], [3, 3, 5, 4], [1, 5, 4, -2], [7, 4, -2, 2]], dtype=float)

eigenvalues_qr, eigenvectors_qr = metodo_qr(A, tol=1e-7)
eigenvalues_numpy, eigenvectors_numpy = LA.eig(A)
```
